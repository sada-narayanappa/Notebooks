## MCMC (Markov Chain Monte Carlo) Methods

### Markov chain

* A probablistic state diagram is a markov chain.
* A Markov Chain is **irreducible** if any state can be reached by any other state
* A state *s* is **recurrent** if starting at that state, there is a path to return to it; otherwise *s* is called transient.

State transitions can be written as transition matrix. For example: consider a Markov Chain of 2 states as shown in *B* and initial state *A*,

$
 A_{1 \times 2} = \begin{bmatrix}
0.55 & 0.45
\end{bmatrix} 
\\ B_{2 \times 2} = \begin{bmatrix}
0.7 & 0.3\\\\0.1 & 0.9
\end{bmatrix} 
$
