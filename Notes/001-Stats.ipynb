# Statistics Quick review

**Construct**: A thing that is difficult to measure; For example - happiness, pain, memory etc.

**Operational Definition**: A methodology to measure construct.
For example: We may decide to measure memory by number of things one could remember shown in a short period 

**Population**: is all the individuals in a particular study. 
For example: if we are studying effect of a drug on Cancer, then all Cancer patients could be in this population

**Sample**: Subset of the the population selected in some way (mostly randomly)

$\mu$ - represents the population average - this is also called as *population parameter*
$\overline{x}$ - is used to represent the average value of a sample - also referred to as *statistic* 

**Parameter Vs/Statistic**: *Parameter* is to population as *statistic* is to sample

**Independent Variable**. The independent variable of a study is the
variable that experimenters choose to manipulate; it is usually plotted along the x-axis of a
graph.

**Dependent Variable.** The dependent variable of a study is the variable
that experimenters choose to measure during an experiment; it is usually plotted along the
y-axis of a graph.

**Treatment Group.** The group of a study that receives varying levels of
the independent variable. These groups are used to measure the effect of a treatment.

**Control Group.**  The group of a study that receives no treatment. This
group is used as a baseline when comparing treatment groups.

**Placebo.** Something given to subjects in the control group so they think
they are getting the treatment, when in reality they are getting something that causes no effect
to them. (e.g. a Sugar pill)

**Blinding.** Blinding is a technique used to reduce bias. Double blinding
ensures that both those administering treatments and those receiving treatments do not know
who is receiving which treatment.

### Types of Study

One could conduct ** Observational Study ** by analyzing the collected data. This reveal some interesting patterns or properties such correlation of variables etc.

**Surveys ** are inexpesive ways of collecting data through public opinions

**Experiments** (Sometimes called controlled experiements) are conducted to validate a hypothesis or to validate **causations**.

### Experimental Study example:

Lets say you want to test the effect of a sleeping medication, one could follow the procedure outlined below:

0. State your Hypothesis - decide - null hypothesis; one tailed/two tailed; $\alpha$ level (5%?)
1. Choose randomly two groups of people, each group consisting of 25 people in each
2. Call the first group **control** who get a placebo 
3. Second group are experimenal group who gets the real medication
4. They are given the medication at the same time, same location in a sleep lab
5. The observers write the quality of sleep from both the groups 
6. The results are evaluated after two week period to test if medication had an effect.

In this experiement, the groups were not told who is in **contol** (who got placebo) and who are in **Experimental group**.
Such studies are known as "blind study"

If the observers also not know the distinction of who are in control or otherwise, then the study is called ** Double blind study**. Sometimes it is necessary to not to disclose such information because such knowledge could influence the results..

## Hypothesis Testing

Hypothesis testing (or experiement) is done is test the effect of some treatment.

For example:
One could test the effect of sleeping pill. Now, if we can make observational study on people using the drug and test if those people are sleeping better. This type os study can show correlation but still it is unclear if the drug is **causing it**.

As we described in the experiemental section, we choose two groups 
1. Control 
2. Treatment group
But the first step is to state the hypothesis

Hypothesis is stated as Null Hypothesis as showing no effect. Null hypothesis $H_0$.   
<br/>
If we disprove the *Null* hypothesis by showing the probablity of seeing those effects (by observing population parameters) are very low, then we can reject *Null* hypothesis in favor of Alternative hypothesis $H_a$ that states the drug has an effect.

Suppose we know the population mean $\mu$ and standard deviation $\sigma$; 
<br/>
If we take sample of size n, give them treatment and measure their average $\bar{x} $ and standard deviation = $\frac{\sigma}{\sqrt{n}}$

We compute the $z$-score:
$$z = \frac{\bar{x} - \mu}{ \frac{\sigma}{\sqrt{n}}}$$
and test if the z-score is in the critical region.

Here is a complete example.

**Null Hypothesis**: $H_0$ is: SONG HAS NO EFFECT

**Alt Hypothesis**: $H_a$ is: SONG HAS Some EFFECT (Notice two tailed test)



By running a hypothesis test on a sample of $n=30$, we found the sample mean 
$$\bar{x} = 8.3 \\
n = 30  \\
$$

Question is - does the song had an effect?
First, we compute the $Z-score$ and compare to see if it falls in the critical region. 

$$ Z = \frac{8.3 - 7.47}{\frac{2.41}{\sqrt{30}}} = 1.89
$$

All of the following statements are equivalent
* $Z$ is less than -ve $z$ or greater than +ve $z$ 
* the $p$ value falls in the critical region

Now, we see that $Z = 1.89$ is not in the critical region. 
<br/>
The probability of getting this result is $1 - .9706 = 0.0294 > 0.025$
<br/> 
We found $0.9706$ by looking in Z-table corresponding to Z-score

Therefore, we cannot reject the null hypothesis; In other words, we got this chance.
