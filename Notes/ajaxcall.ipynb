```python
%%html
<script>
$("site").css('height', '100%')

function ToggleCode(code_show){
    showHide = (code_show) ? "block": "none"
    $('div.cell.code_cell').each(function(i){
        if (i >= 0){
            console.log(this)
            $(this).css('display', showHide);
            return
        }
        $(this).css('display', 'none');
    });    
}

code_show=false;
function ToggleEle(element) {
    if (code_show){
        $(element).css('display', 'block');
    } else {
        $(element).css('display', 'none');
    }
    ToggleCode(code_show)
    
    code_show = !code_show
    
}

function Ntoggle(){
    ToggleEle('#header')
}
$('body').append('<input type=button value="--" style="position: fixed;right:30;bottom:0" z-index: 999999 ; onclick="Ntoggle()" />') 
$('#site').css('height', '100%');

</script>


```
