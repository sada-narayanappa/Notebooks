```python
def stack(xo, lags=0, shift=0):
    if ( shift <0 or lags <0 or shift+lags >= len(xo)):
        return xo;
    x = xo[shift:]
    y = np.array(x[lags:])
    ll = len(y)
    a=[]
    for i in range(0,lags):
        n=x[i:i+ll];
        a.append(n)
    #a.append(r)
    X = np.vstack(a).transpose()
    print("Orig: {}, \n y: {} \n X:\n {}".format(xo, y, X))
    return y.astype(float), X.astype(float)

a=np.array(range(0,10)).astype(float)
#print(a, a[1:], a[:-2], a[:-3] )
y, X = stack(a,1,0)
#print(X)
r = sm.OLS( y, X).fit()
print (r.aic, r.nobs, r.resid, r.rsquared_adj, r.params)
print(r.summary())
```
