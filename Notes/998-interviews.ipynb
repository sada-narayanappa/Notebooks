### Statistical Questions interview questions for data scientist 

* ** What is P-value **
P Values. The P value, or calculated probability, is the probability of finding the observed, or more extreme, results when the null hypothesis ($H_0$) of a study question is true – the definition of 'extreme' depends on how the hypothesis is being tested. A p-value less than alpha-level (a predetermined level) shows probability of making a mistake of rejecting the null-hypothesis.

* A small p-value (typically ≤ 0.05) indicates strong evidence against the null hypothesis, so you reject the null hypothesis.

* A large p-value (> 0.05) indicates weak evidence against the null hypothesis, so you fail to reject the null hypothesis.

* p-values very close to the cutoff (0.05) are considered to be marginal (could go either way). Always report the p-value so your clients can draw their own conclusions.

### Correlation 

** Correlation Vs Causation **

Correlation is relationship of one variable to another.

You could see this via scatter plot or scatter_matrix in python which shows the matrix of plots of one to other

Pearson's $r$ ($r_p$) and Spearman's r ($r_s$) (both denoted as $\rho$ for population is another way to test the correlation between two variables). 

Pearson's gives linear relationship and Spearman's gives monotonic relationship.
If they differ look for outliers. 

$$ r = \frac{Covariance(x,y)}{S_x,S_y} $$

######  What does Correlation Coefficient (Pearson's r) says   

Correlation coefficient gives the strength of the relationship and ranges between [-1,1]. Closer to 1 means stronger the relationship. Positive relationship says, an increase in one, increases the value of the other. 

Once you get the $r$ value, we do hypothesis testing to get the $p$ value to see if it is meaningful; i.e. to test the null-hypothesis the "there is no relationship"

Furthermore, one should check with the domain knowledge if the relationship is meaningful.

For ex: what would say, if you found the correlation of number of eggs in a fridge to number of desktop icons? 

######  Why and what is $R^2$ - Coefficient of determination    

** Coefficient of Determination ($r^2$) ** is the percentage of the variation in the dependent variable (y) that can be explained by variation in the independent variable.

Suppose in a study conducted, age and salary the $R^2$ value was 31%. It means 31% of the variations in salary depends on age. Other 69% may be explained by other factors that are not in the system - perhaps education level, domain where they work, part-time or full-time etc.

> Bootstrap or .632 bootstrap

What is the chance that a particular instance will not be picked for the training set? It has a 1/n probability of being picked each time and so a 1 – 1/n probability of not being picked. Multiply these probabilities together for a sufficient number of picking opportunities, n, and the result is

$$ 1-\frac{1}{n} ≈ e^{−1} =0.368 $$

where e is the base of natural logarithms, 2.7183 (not the error rate!) This gives the chance of a particular instance not being picked at all. Thus, for a reasonably large dataset, the test set will contain about 36.8% of the instances and the training set will contain about 63.2% of them (now you can see why it’s called the 0.632 bootstrap). Some instances will be repeated in the training set, bringing it up to a total size of n, the same as in the original dataset.

* ** What does it mean to control for the variables in the model? ** It means that when you look at the effect of one variable in the model, you are holding constant all of the other predictors in the model. You explain the effect that changes in one predictor have on the response without having to worry about the effects of the other predictors. In other words, you can isolate the role of one variable from all of the others in the model. And, you do this simply by including the variables in your model.
